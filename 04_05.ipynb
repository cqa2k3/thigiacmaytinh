```python
# import thu vien
from PIL import Image
from imgtools import *
```

```python
# duong dan anh
my_path = 'D:/hinhanh/anh1.jpg'
```

```python
# doc anh
img  = load_image(my_path)
```

```python
display(img)
```

```python
# dinh vi khu vuc cat 
region = (100, 100, 150,200)
# cat vung anh
cropped_image = img.crop(region)
```

```python
display(cropped_image)
```

```python
# dan
# xac dinh vi tri can dan
paste_positon = (150, 150)
img.paste(cropped_image, paste_positon)
display(img)
```

```python
display(img)
```

```python
# xoay ảnh
rotated_image = img.rotate(90)
display(rotated_image)
```

```python
rotated_image = rotated_image.rotate(90)
display(rotated_image)
```

```python
# Transpose ảnh
transposed_image = img.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```

```python
# Transpose ảnh
transposed_image = transposed_image.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```
